prediction of Online Shoppers Purchase Intentions.
1.Name the product most liked and purchased
2.identify the special days when most products are purchased
3.list the top 5 special days of purchase and type of visitors.
4.Identify the months when most purchases happened.
5.what the max and min revenue generated by the online shoppers.

```python
import pandas as pd

# Load dataset
data = pd.read_csv('online_shoppers_intention.csv')

# Most liked and purchased product
most_liked_product = data['ProductRelated'].mode()[0]

# Special days with purchase counts
special_days_counts = data['SpecialDay'].value_counts()

# Top 5 special days
top_special_days = special_days_counts.head(5)

# Categorizing visitors
top_special_days_with_visitors = data.groupby(['SpecialDay', 'VisitorType']).size().unstack(fill_value=0)

# Months with highest purchases
months_counts = data['Month'].value_counts()

# Revenue statistics
max_revenue = data['Revenue'].max()
min_revenue = data['Revenue'].min()

# Display results
print(f"Most Liked Product: {most_liked_product}")
print("Top 5 Special Days and Purchase Counts:\n", top_special_days)
print("Top Special Days by Visitor Type:\n", top_special_days_with_visitors)
print("Months with Highest Purchases:\n", months_counts)
print(f"Max Revenue: {max_revenue} \nMin Revenue: {min_revenue}")
```
